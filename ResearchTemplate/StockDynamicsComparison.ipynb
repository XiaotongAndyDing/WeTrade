```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

module_path = os.path.abspath(os.path.join('..'))

if module_path not in sys.path:
    sys.path.append(module_path+"\\Source")

from Source.Market import StockGeometricBrownianMotion, StockMeanRevertingGeometricBrownianMotion, StockTrendingGeometricBrownianMotion
```

Geometric Brownian Motion with drift 0 and annualized volatility 20%

```python
for _ in range(3):
    stock_gbm = StockGeometricBrownianMotion('stock_gbm_test', 100, 0, 0.2 / np.sqrt(252))  
    # annualized stock vol 20%, 252 days per year
    stock_gbm.mark_current_value_to_record(0)
    
    for time in range(1, 252):
        stock_gbm.evolve(time)
        stock_gbm.mark_current_value_to_record(time)
    
    plt.plot(stock_gbm.price_record.keys(), stock_gbm.price_record.values())
plt.xlabel('time (day)')
plt.ylabel('stock price')
```

Mean-Reverting Geometric Brownian Motion with drift 0, annualized volatility 20%, equilibrium price=100 and mean reversion speed 0.001
